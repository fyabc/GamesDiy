# 三国杀 收益论

## 符号约定

1. 某种卡牌在牌堆中的频率用 $[牌名]$ 表示，如 $[杀], [闪]$ 等
2. 游戏模式：$M$
   1. 身份模式（$M \in R(角色数, (主公数,忠臣数,反贼数,内奸数))$）：身份标记$\rm Role \in \{主公,忠臣,反贼,内奸\}$

    例子：
    
    - $M = \rm{"1v1"}$ 1vs1模式
    - $M = R(5, (1, 1, 2, 1))$ 标准五人身份模式
    - $M = R(8, (1, 2, 4, 1))$ 标准八人身份模式
    - $M = R(8, (1, 2, 3, 2))$ 八人双内奸身份模式
    
3. 游戏变量
    1. 角色数：$N$
    2. 牌堆：$\rm Deck$
    3. 玩家$i$：$P_i$
    4. 玩家$i$的**体力值**：$H_i$或简写为$H$。
    5. 玩家$i$的**手牌数**：$C_i$或简写为$C$（为简便，下文不再写出下标形式）。
    6. 攻击范围：$\rm Range$
    7. 装备区：$\rm Equip$
    8. 体力上限：$L_H$
    9. 手牌上限：$L_C$
    10. 更多 ...

    > 某个变量$X$的变化以$\Delta X$表示

4. 卡牌收益与技能收益
   1. 伤害值：$d = \Delta H$
   2. 事件$e$的概率：$p_e$
   3. 一张卡牌$c$的基础价值：$B_c$或简写为$B$
   4. 一张卡牌$c$的**期望**使用收益：$U_c$或简写为$U$
   5. 一张卡牌$c$的**期望**净收益：$V_c = U_c - B_c$或简写为$V$
   6. 一张卡牌$c$的**实际**使用收益$U'_c$和**实际**净收益$V'_c$

## 卡牌价值讨论

### 概述

1. 每张卡牌$c$有使用收益$U$，定义为使用该牌造成体力值变化（伤害）的期望值：$U = E[\Delta H] = E[d]$。
2. 每张卡牌$c$有基础价值$B$，这是不考虑卡牌内容的价值（即卡牌在未使用情况下的价值）。该数值对所有卡牌都相同。
    1. 对于标准“一血两牌”模型，$B = 0.5$。故**本文中的收益需要乘以2得到通常的收益论中的数值**。
3. 因此，使用一张卡牌$c$的净收益$V = U - B$（需要扣除使用此牌后造成的基础价值损失）。

### 一些基本数值

1. 使用【杀】命中目标的概率为$p_{命中}$
1. 使用一张锦囊牌，其中一个目标被【无懈可击】抵消的概率为$p_{无懈}$
1. 全牌表的卡牌分布$D_{\rm all}$（简称“基本分布”）
1. 牌堆中所有卡牌的平均净收益：$\overline{V} = \frac{1}{|\rm Deck|}\sum_{c \in \rm{Deck}}V_c$
1. 所有装备区中的牌的平均使用收益$\overline{U_{\rm 装备区}}$
1. 所有判定区中的牌的平均使用收益$\overline{U_{\rm 判定区}}$
1. 从$M$张手牌中主动弃置$N$张的平均净收益：$\overline{V_{主动弃置手牌}(m, n)}$


### 基本牌

1. 普通【杀】
    
    1. 以$p_{命中}$的概率命中目标，收益为$U = 1$
    2. 以$1 - p_{命中}$的概率被【闪】抵消，收益为$U = B$（消耗掉一张【闪】）
    3. 净收益：$V_{普杀} = p_{命中} \cdot 1 + (1 - p_{命中}) \cdot B - B = p_{命中}(1 - B)$
    4. 估算1：$p_{命中} \approx \frac{[闪]}{[杀]}$，故$V_{普杀} \approx \frac{[闪]}{[杀]} \cdot (1 - B)$
    5. 估算2：在标准和军争牌堆中，闪杀比$\frac{[闪]}{[杀]}$约为$0.5$，故$V_{普杀} \approx 0.5(1 - B)$
    6. 方差：$V'_{普杀} \in [0, 1 - B]$

### 锦囊牌

1. 【无中生有】

    1. 净收益：$V_{无中生有} = (1 - p_{无懈}) \cdot 2B + p_{无懈} \cdot B - B = B(1 - 2p_{无懈})$
    2. 方差：$V'_{无中生有} \in [-B, B]$

2. 【过河拆桥】

    1. 设目标为装备区的牌和判定区的牌的概率分别为$p_{拆装备区}$和$p_{拆判定区}$
    2. 净收益：
    
        \begin{equation}
        \begin{split}
        V_{过河拆桥} & = (1 - p_{无懈}) \cdot (p_{拆装备区} \overline{U}_{装备区} + p_{拆判定区} \overline{U}_{判定区} + (1 - p_{拆装备区} - p_{拆判定区}) \cdot B) + p_{无懈} \cdot B - B \\
        & = (1 - p_{无懈}) \cdot (p_{拆装备区} (\overline{U}_{装备区} - B) + p_{拆判定区} (\overline{U}_{判定区} - B))
        \end{split}
        \end{equation}

3. 【兵粮寸断】

    1. 净收益：$V_{兵粮寸断} = (1 - p_{无懈}) \cdot 0.75 \cdot 2B + p_{无懈} \cdot B - B = (1 - p_{无懈}) \cdot 0.5B$
    2. 方差：$V'_{兵粮寸断} \in [-B, B]$

4. 【乐不思蜀】
    1. 净收益（较难估算）：$V_{乐不思蜀} = (1 - p_{无懈}) \cdot 0.75 \cdot (回合内使用牌收益 + 回合内技能收益 + 平均弃牌负收益) + p_{无懈} \cdot B - B$
    2. 估算1：所有武将的平均回合内技能收益约为0.75（四血武将回合内收益约为0.5，三血武将回合内收益接近1.0）
    3. 估算2：回合内使用牌的收益约为$2\overline{V}$（假设每回合摸两张牌）
    4. 估算3：平均弃牌负收益约为$2B$（假设弃置两张牌，弃置更多牌的情况和可选择弃置价值最低的牌的情况相抵消）
    5. 故$V_{乐不思蜀} \approx (1 - p_{无懈}) \cdot (0.75 \cdot (2\overline{V} + 0.75 + 2B) - B) = (1 - p_{无懈}) \cdot (1.5\overline{V} + 9/16 + 0.5B)$

## 技能价值讨论

1. 主要指标：每轮平均收益
2. 白板武将的每轮收益

    1. 摸牌：$2B$
    2. 用牌：$2\overline{V}$
    3. 总收益：$2\overline{U} = 2\overline{V} + 2B$

3. 主动/被动弃牌的（负）收益
    1. 基本问题：从总共可选的$m$张手牌中弃置$n$张牌（$1 \le n \le m$）
    2. 解法：确定可选牌分布和选牌方式，计算负收益期望值
    2. 假设1：主动弃牌时，玩家总会弃置可选牌中价值最低的$n$张牌；被动弃牌则为随机选择（有可见的装备牌等更高价值牌优先）
    3. 假设2：“主动弃置手牌”条件下，可选牌分布约等于基本分布$D_{\rm all}$
